```python
bidder_info = df.groupby("bidder").agg( total_bids = ("bidder", "size"), wins = ("win_count","sum"))

bidder_info["win_rate"] = bidder_info["wins"]/bidder_info["total_bids"]
```

```python
winner_bidder = bidder_info["win_rate"].idxmax()
```
